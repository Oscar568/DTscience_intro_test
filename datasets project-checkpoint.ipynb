```python
import pandas as pd
data_companies = pd.read_csv('S&P 500 Stocks/sp500_companies.csv')
data_companies
```

```python
#Cleaning data with specific methods
df1 = pd.DataFrame(data_companies, columns = ['Currentprice','Marketcap','Revenuegrowth'])
#to display all data
pd.set_option('display.max_rows', None)

df1['difference'] = df1['Currentprice']- df1['Revenuegrowth']
df1.sort_values('difference', ascending = False)
```

```python
#seaborn is data visualization library on matplotlib
import seaborn as sns
sns.set()

graph = sns.scatterplot(x= 'Currentprice', y = 'Revenuegrowth', data=df1)
graph
```
